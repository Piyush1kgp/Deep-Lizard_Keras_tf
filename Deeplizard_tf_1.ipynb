```
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import  MinMaxScaler
```

```
train_lables = []
train_samples = []
```

We create our own random dataset with theme as: 1. 2100 patients went to clinical trail with age 13-100 out of which 95% with aged more than 65 yrs experienced side effects while 5% for aged less than 65. Out of total pool ot participants, half of total population is aged less than 65 yrs.

```
for i in range(50):
  #5% that experienced side effects
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_lables.append(1)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_lables.append(0)

for i in range(1000):
  #5% that experienced side effects
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_lables.append(0)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_lables.append(1)
  
```

```
for i in train_samples:
  print(i)
```

```
for i in train_lables:
  print(i)
```

```
#Now converting the list to numpy array
train_samples = np.array(train_samples)
train_lables = np.array(train_lables)
train_lables, train_samples = shuffle(train_lables, train_samples)
```

```
train_samples
```

```
train_samples.reshape(-1,1) #used to convert 1-d to 2-d array so that it can be used by function for transformation
```

```
scaler = MinMaxScaler()
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```
scaled_train_samples
```

Now utilising the Keras API

```
import tensorflow as tf
```

```
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
#Above ones will be used while we build our model and lower ones while we train the model
from tensorflow.keras.optimizers import Adam 
from tensorflow.keras.metrics import categorical_crossentropy
```

```
#To check whether tf has identifies the GPU and as well as enable memory growth
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("Num GPUs available",  len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

```
#let's build a Sequential model which can be defined as linear stack of layers
model = Sequential([
                    Dense(units = 16, input_shape=(1,), activation = 'relu'),
                    Dense(units = 32, activation = 'relu'),
                    Dense(units = 2, activation = 'softmax')
])
```

```
"""for the dense layer with 16 units: first hidden/ second layer including first as input layer. Units are also called nodes/neurons
for the dense layer with 2 units: it depicts it as final layer or output layer. 2 units corresponds to the possibility of 2 output classes
"""
```

```
model.summary()
```

```
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics =['accuracy']) # This prepares the model for training
```

```
#model.fit(x = scaled_train_samples, y = train_lables, batch_size=15, epochs=29, verbose=2, shuffle= True)
```

```
model.fit(x = scaled_train_samples, y = train_lables, validation_split = 0.1, batch_size=10, epochs=30, verbose=2, shuffle= True)
# This function helps to get a separate set of validation set that can be used to identify how general our model is
# We need data to be shuffled before applying the validation split, using the above function with shuffle = True , it will be executed before validation_split
# Hence, here the shuffle we made in starting codes proved to be helpful
```

```
# Hence, our model is generalizing pretty well and it's not overfitting the training dataset
```
